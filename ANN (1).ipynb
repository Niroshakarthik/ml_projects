Introduction


In this project, I build an Artificial Neural Network (ANN) to predict customer churn using the Churn Modelling dataset. Customer churn refers to the loss of customers, and predicting churn is crucial for businesses to implement strategies to retain customers. The Churn Modelling dataset contains information on customers of a bank, including their demographics, account information, and whether they have churned.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
X=df.iloc[:,3:13]
X.head()
```

```python
y=df.iloc[:,13]
y.head()
```

Encoding the categorical data

Using Label encoder to convert the categorical data into numerical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
X.head()
```

```python
X["Geography"]=le.fit_transform(X["Geography"])
X.head()
```

```python
X["Gender"]=le.fit_transform(X["Gender"])
X.head()
```

To deploy neural network we need to split the data into train test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

Creating own ANN model
Listing the steps:
1.Randomly initialize the weights to small numbers close to 0 but not 0

2.inputing the first observation of the dataset in the input layer each feature in one input node.

3.Forward propagation from left to right the neurons are activated in a way that the impact of each neuron activation is limited by the weights.propagate the activation until getting the predicted result

4.Comparing the predicted result with the actual result. Measuring the generated error

5.back propagation: From right to left error is back propagated.Update the weights according to how much they are responsible for the error.Learning rate  tells us how much we update the weight.

6.Repeating the step 1 to 5 and update the weight after each observation(Reinforcement learning)

7.When the whole training data is passed through the ANN  That completes an Epoch.

Importing Keras library and packages

```python
import keras
# for building the neural network layer by layer
from keras.models import Sequential
```

```python
# to randomly initailize the weights to small numbers close to 0 (But not 0)
from keras.layers import Dense
```

Intialising the ANN :
There are two ways to initailise the deep learning model
1.defining each layer one by one
2.defing a Graph

Adding the Input layer and 1st hidden layer

```python
classifier=Sequential()
```

```python
#defining the input layer
classifier.add(Dense(5,kernel_initializer='glorot_uniform',activation='relu',input_dim=10))
```

```python
classifier.add(Dense(5,kernel_initializer='uniform',activation='sigmoid'))
classifier.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

compiling the ANN

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#Training the model with data
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

Predicting the Test result

```python
 y_pred=classifier.predict(X_test)
```

```python
print(y_pred)
```

```python
churn=[]
churn_val=[]
for i in y_pred:
  if i>0.5:
    churn.append("YES")
    churn_val.append(1)
  else:
    churn.append("NO")
    churn_val.append(0)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,churn_val)
print(cm)
print(accuracy_score(y_test,churn_val))
```

This model when trained on the train data and made test on test data gave accuracy of around 84% In both cases
